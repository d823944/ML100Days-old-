```python
app_train.describe()
```

```python
print(f'Mean of AMT_INCOME_TOTAL = {app_train["AMT_INCOME_TOTAL"].mean():.0f}')
print(f'STD of AMT_INCOME_TOTAL = {app_train["AMT_INCOME_TOTAL"].std():.0f}')
```

```python
AMT_INCOME_TOTAL = np.log10(app_train['AMT_INCOME_TOTAL'])
AMT_INCOME_TOTAL.plot(kind='hist', bins=50)
xmin = int(np.floor(AMT_INCOME_TOTAL.min()))
xmax = int(np.ceil(AMT_INCOME_TOTAL.max()))
plt.xlim(xmin, xmax)
plt.xticks(np.arange(xmin, xmax+1), [f'{10**i:.0e}' for i in range(xmin, xmax+1)])
plt.xlabel('AMT_INCOME_TOTAL')
```
