1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection
2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?


#### Answers
- 1. Lasso's loss function: $argmin(y-wx)^2+\alpha |w|$. The objective is to find $w$ as small as possible to minimize the loss, therefore, it will lead to zero weights for those insignification features by nature.

- 2. Yes. With multicolinearity, a small change in a single data point can wildly swing the coefficient estimates. Using Ridge Regression, the objective is to minimize $argmin(y-wx)^2+\lambda w^2$ and will penalize large weights.
